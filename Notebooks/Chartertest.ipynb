```python
import pandas as pd
```

```python
colo = pd.read_csv("../Data/Colocation2025 .csv")
colo.columns
colo['District'] = colo[' District']
```

```python
colodistrict = colo.groupby(' District').size().reset_index(name='count')
colodistrict.columns = ['District', 'count']
colodistrict.dtypes
```

```python
colodistrictenrollment = colo[['District', 'Total Enrollment']]
```

```python
colodistrictenrollment  = colodistrictenrollment.groupby('District')['Total Enrollment'].sum().reset_index()
```

```python
colodistrictenrollment 
```

```python
coloc = pd.merge(colodistrictenrollment, colodistrict, on ='District')
```

```python
coloc.to_csv("../Output/coloc.csv")
```

```python
colo['School Name'] = colo['School Name'].astype(str)
colo['success'] = colo['School Name'].str.contains('Success Academy')
colosuccess = colo[colo['success'] == True].reset_index(drop=True) 
notsuccess = colo[colo['success'] == False].reset_index(drop=True) 
colosuccess.to_csv("../Output/colosuccess.csv")
notsuccess.to_csv("../Output/notsuccess.csv")
```

```python
successdistrict = colosuccess.groupby('District').size().reset_index(name='count')
successdistrict.to_csv("../Output/successdistrict.csv")
successborough = colosuccess.groupby('Borough').size().reset_index(name='count')
successborough.to_csv("../Output/successborough.csv")
successtype = colosuccess.groupby('School Type').size().reset_index(name='count')
successtype.to_csv("../Output/successtype.csv")
```

```python
notsuccessdistrict = notsuccess.groupby('District').size().reset_index(name='count')
notsuccessdistrict.to_csv("../Output/notsuccessdistrict.csv")
notsuccessborough = notsuccess.groupby('Borough').size().reset_index(name='count')
notsuccessborough.to_csv("../Output/notsuccessborough.csv")
notsuccesstype = notsuccess.groupby('School Type').size().reset_index(name='count')
notsuccesstype.to_csv("../Output/notsuccesstype.csv")
```
